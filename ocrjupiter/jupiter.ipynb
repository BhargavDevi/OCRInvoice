```python
%pip install matplotlib

import easyocr
from matplotlib import pyplot as plt
import numpy as np
import cv2
```

```python
Image_path='WhatsApp Image 2025-06-24 at 14.32.35.jpeg'
```

```python
reareader = easyocr.Reader(['en'], gpu=False)  
result = reareader.readtext(Image_path)
result
```

```python
import re
import json

# Replace this with your actual OCR data (output you shared earlier)
ocr_data = result

# Convert to plain list of text lines (simulate reading order)
text_blocks = [entry[1] for entry in ocr_data]

# Join for regex-based searches
full_text = "\n".join(text_blocks)

# Basic patterns (can be enhanced)
invoice_number = None
invoice_date = None
gst_number = None
total_amount = None
company_name = None

# Try to extract fields using regex
for i, text in enumerate(text_blocks):
    # Invoice number
    if re.search(r'invoice\s*no', text, re.IGNORECASE):
        if i + 1 < len(text_blocks):
            invoice_number = text_blocks[i + 1]

    # Date (look for keywords and date patterns)
    if re.search(r'date', text, re.IGNORECASE):
        if i + 1 < len(text_blocks):
            date_candidate = text_blocks[i + 1]
            match = re.search(r'\d{1,2}[/-]\d{1,2}[/-]\d{2,4}', date_candidate)
            if match:
                invoice_date = match.group()

    # GST number (Indian format)
    match = re.search(r'\b\d{2}[A-Z]{5}\d{4}[A-Z]{1}[A-Z\d]{1}[Z]{1}[A-Z\d]{1}\b', text)
    if match:
        gst_number = match.group()

    # Total amount (find after "Total" or "Amount")
    if re.search(r'\btotal\b|\bamount\b', text, re.IGNORECASE):
        for j in range(i+1, min(i+3, len(text_blocks))):
            amt_match = re.search(r'₹?\s?\d+[.,]?\d*', text_blocks[j])
            if amt_match:
                total_amount = amt_match.group()
                break

    # Company name (first large caps string or near top)
    if not company_name and text.isupper() and i < 5:
        company_name = text

# Compile results
invoice_data = {
    "invoice_number": invoice_number,
    "invoice_date": invoice_date,
    "company_name": company_name,
    "total_amount": total_amount,
    "gst_number": gst_number,
}

# Print as JSON
print(json.dumps(invoice_data, indent=2))
```
